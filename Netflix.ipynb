- Both types show a sharp increase in releases starting in the 2000s, with a peak near 2020 â€” likely driven by streaming platforms and global content expansion.
- The smoothed lines help visualize the upward trend and show that while Movies have always been more frequent, TV Shows are catching up in recent years.


***Histogram: Release Year Distribution by Rating***
This chart shows how the number of movies released each year varies across different content ratings, giving us a layered view of how rating categories evolved.

Key Observations

- Post-1980s surge: Movie releases increased dramatically after the 1980s, peaking in the 2010s.

- PG, PG-13, and R ratings are consistently present, especially from the 1990s onward â€” typical for mainstream cinema.
- Youth ratings like TV-Y, TV-G, and G appear sporadically, suggesting fewer releases aimed at younger audiences.
- Rare ratings (NC-17, UR, TV-Y7-FV, NR) show minimal presence across all years.
